## Распределение Лапласа

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
alfa = 1 #Коэффициент масштаба
beta = 0 #Коэффициент сдвига
size = 1000
laplace = sts.laplace(loc=beta, scale=alfa)

# Зададим 1000 случайных значений, подчиненных распределению Лапласа
sample = laplace.rvs(size)

print ('\033[4m' + 'Theoretical', '\033[0m')
x = np.linspace(-5,5,100)
pdf = laplace.pdf(x) #Значение плотности
plt.hist(laplace.rvs(size), color='purple', density = True) #гистограмма по 1000 полученным случайным значениям
plt.plot(x, pdf, label='theoretical pdf') #График плотности
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()

# функция 1000 раз генерирует выборку объема n, подчиненную распределению Лапласа, вычисляет среднее значение,
# и добавляет его в list_x
def sample_laplace(n):
    list_x = []
    for i in range(1000):
        list_n = laplace.rvs(n)
        list_x.append(round(np.mean(list_n),3))
    return list_x

# функция формирования гистограммы выборки и графика плотности соответствующего нормального распределения с параметрами:
# среднее распределения Лапласа равняется его параметру бета
# дисперсия равняется выражению (2/альфа^2)
def graph_laplace(list_x, n):
#   гистограмма выборки
    plt.hist(list_x, bins = 'auto', color='purple', density = True, alpha=0.6)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
#   график плотности
    norm_rv = sts.norm(beta, ((2/(alfa**2))/n) ** (0.5))
    x = np.linspace(-3, 3, 100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='laplace pdf', alpha=1)
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
```

```python
#Определим объем выборок:
choice = (5,10,100,1000)
for n in (choice):
    for i in range(1):
        list_x = sample_laplace(n)
        print ('\033[4m' + 'Объем выборки:', '\033[1m', n, '\033[0m')
        graph_laplace(list_x, n)
```
